```python
def double_char(s):
    l=len(s)
    n=""
    for i in range (0,l):
        n+=s[i:i+1]+s[i:i+1]
    print(n)
double_char("String")     
double_char("Hello World") 
double_char("1234!_ ")
```

```python
#Question2
def reverse(inp):
    if(inp==True):
        print("False")
    elif (inp==False):
        print("True")
    else:
        print("boolean expected")
reverse(True) 
reverse(False) 
reverse(0) 
reverse(None)
```

```python
#question3
def num_layers(a):
    res = 0.5
    for _ in range(a):
        res *= 2
    return res / 1000 
print(num_layers(1))
print(num_layers(4))
print(num_layers(21))
```

```python
#question4
def index_of_caps(s):
    l=[]
    for i in s:
        if (i.isupper()):
            l.append(s.index(i))
    print(l)
index_of_caps("eDaBiT") 
index_of_caps("eQuINoX") 
index_of_caps("determine")
index_of_caps("STRIKE") 
index_of_caps("sUn") 
```

```python
#question5:
def find_even_nums(n):
    even= [i for i in range(1,n+1) if i%2==0]
    return even
print(find_even_nums(8))
print(find_even_nums(4))
print(find_even_nums(2))
```
